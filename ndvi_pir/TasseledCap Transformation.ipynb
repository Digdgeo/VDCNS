```python
%matplotlib inline 

import os
import rasterio
import subprocess
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = r'I:\ndvi_PIR\ndvis'
```

```python
'''Brightness = 0.3037*OLI 2 + 0.2793*OLI 3 + 0.4743*OLI 4 +
0.5585*OLI 5 + 0.5082*OLI 6 + 0.1863*OLI 7 (2)
Greenness =-0.2848*OLI 2 – 0.2435*OLI 3 – 0.5436*OLI 4 +
0.7243*OLI 5 + 0.0840*OLI 6 - 0.1800*OLI 7 (3)
Wetness = 0.1509*OLI 2 + 0.1973*OLI 3 + 0.3279*OLI 4 +
0.3406*OLI 5 - 0.7112*OLI 6 - 0.4572*OLI 7'''
```

```python
#Vamos a hacer un diccionario con cada una de las bandas de las Landsat para hacer los calculos
#d = {b1:'pathtpb1', b2: pathtob2....}
def read_raster(raster):
    
    with rasterio.open(raster) as src:
        R = src.read()
        
    return R

for i in os.listdir(path):
    
    carpeta = os.path.join(path, i)
    
    if os.path.isdir(carpeta):
        
        d = {}
        
        for banda in os.listdir(carpeta):
            
            if banda.endswith('.TIF') and not 'ndvi' in banda:
                b_name = banda[-6:-4]
                d[b_name] = os.path.join(carpeta, banda)
        
        print(len(d.keys()))
        if len(d.keys()) == 6:
            
            print('Haciendo escena', i)
                        
            Brightness = 0.3037 * read_raster(d['b1']) + 0.2793 * read_raster(d['b2']) + 0.4743 * read_raster(d['b3']) + \
                        0.5585 * read_raster(d['b4']) + 0.5082 * read_raster(d['b5']) + 0.1863 * read_raster(d['b7'])        
            Greenness = -0.2848 * read_raster(d['b1']) - 0.2435 * read_raster(d['b2']) - 0.5436 * read_raster(d['b3']) + \
                        0.7243 * read_raster(d['b4']) + 0.0840 * read_raster(d['b5']) - 0.1800 * read_raster(d['b7'])        
            Wetness = 0.1509 * read_raster(d['b1']) +  0.1973 * read_raster(d['b2']) + 0.3279 * read_raster(d['b3']) + \
                        0.3406 * read_raster(d['b4']) - 0.7112 * read_raster(d['b5']) - 0.4572 * read_raster(d['b7'])        

            out_B = os.path.join(carpeta, i + '_brightness.TIF')
            out_G = os.path.join(carpeta, i + '_greenness.TIF')
            out_W = os.path.join(carpeta, i + '_wetness.TIF')
            
            
            with rasterio.open(d['b1']) as src:
                banda1 = src.read()
                
            profile = src.meta
            profile.update(dtype=rasterio.float32)
            
            with rasterio.open(out_B, 'w', **profile) as dst:
                dst.write(Brightness.astype(rasterio.float32))
            
            print('Brightness creado de:', carpeta)
            
            with rasterio.open(out_G, 'w', **profile) as dst:
                dst.write(Greenness.astype(rasterio.float32))
            
            print('Greenness creado de:', carpeta)

            with rasterio.open(out_W, 'w', **profile) as dst:
                dst.write(Wetness.astype(rasterio.float32))
            
            print('Wetness creado de:', carpeta)
                
        else:
            
            print('La escena no tenía las 6 bandas')
        
        print('NEXT SCENE!!!!!!!!!')
```
