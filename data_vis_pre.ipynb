```python
%matplotlib inline

import pandas, seaborn
import matplotlib.pyplot as plt

df = pandas.read_csv(r'O:\VDCNS\Curvas_MTN_Historico\Datos_Cota\errores_pre_data.csv', sep=';')

plt.rcParams['figure.figsize']=15,10
seaborn.distplot(df['LIDAR'], bins=20)
```

```python
data = pandas.DataFrame({"MTN": df['MTN'], "TopoR_cv": df['TopoR_Curvas'], "TopoR_p": df['TopoR_Points'], 'Tin_cv': df['TIN_Curvas'], 'Lidar': df['LIDAR']})
seaborn.jointplot(x="TopoR_p", y="TopoR_cv", data=data)
```

```python
seaborn.jointplot(x="TopoR_p", y="TopoR_cv", data=data, kind="kde")
```

```python
seaborn.boxplot(data)
```

```python
seaborn.pairplot(data)
```

```python
with seaborn.axes_style("whitegrid"):
    seaborn.jointplot(x="Lidar", y="TopoR_cv", data=data, kind="hex", color="r")
```
